```python
import psycopg2

# データベース接続情報
host = 'postgis'
port = '5432'
user = 'postgres'
password = 'postgres'
database = 'sandbox'

# PostGISへの接続
connection = psycopg2.connect(
    host=host,
    port=port,
    user=user,
    password=password,
    database=database
)

# クエリの実行
cursor = connection.cursor()

# 浸水リスクの高いエリアの近くの病院を探すクエリ
query = """
SELECT h.name
FROM hospitals_akita_2023_07_18 AS h, flood_areas_akita_2023_07_19 AS f
WHERE ST_DWithin(ST_Transform(h.geom, 4326), ST_Transform(f.geometry, 4326), 1000)
"""

# クエリの実行と結果の取得
cursor.execute(query)
results = cursor.fetchall()

# 結果の表示
for row in results:
    print(row[0])

# コネクションのクローズ
cursor.close()
connection.close()
```
